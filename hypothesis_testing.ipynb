### Use the following contingency table to help answer the question of whether using a Macbook and being a Codeup student are independent of each other.

```python
data_dict = {'Codeup Students': [49, 1], 'Not a Code Student': [20, 30]}
data_dict
```

```python
observed = pd.DataFrame(data_dict, index=['Uses a Macbook', 'Doesn\'t Use a Macbook'])
observed
```

```python
alpha = 0.05
```

```python
stats.chi2_contingency(observed)
```

```python
chi2, p, dof, expected = stats.chi2_contingency(observed)
```

```python
print('observed')
print(observed.values)
print('\nExpected')
print(expected.astype(int))
print('\n----')
print(f'chi^2 = {chi2:.4f}')
print(f'pvalue = {p:.4f}')
```

```python
# 1. Conclude based on above values
# if p < alpha:
#         print('reject the null hypothesis')
# else:
#         print('fail to reject the null hypothesis')
# Answer: reject the null hypothesis
# Conclude: there is a relationship between codeup students and using macbook
```

```python
# 3. Use the data from the employees database to answer these questions:

# Is an employee's gender independent of whether an employee works in sales or marketing? 
# (only look at current employees)
# Is an employee's gender independent of whether or not they are or have been a manager?
```

```python
query = '''
        SELECT * 
        FROM employees
            JOIN dept_emp
                USING (emp_no)
            JOIN departments
                USING (dept_no)
        WHERE to_date > NOW()
'''
```

```python
url = env.get_db_url('employees')
```

```python
employee_db = pd.read_sql(query, url)
employee_db
```

```python
# Is an employee's gender independent of whether an employee works in sales or marketing? 
new_query = pd.crosstab(employee_db.gender, employee_db.dept_name)
new_query
```

```python
new_query_dept = new_query[['Sales', 'Marketing']]
new_query_dept
```

```python
# Is an employee's gender independent of whether or not they are or have been a manager?
```

```python
alpha = 0.05
stats.chi2_contingency(new_query_dept)
```

```python
chi2, pvalue, dof, expected = stats.chi2_contingency(new_query_dept)
```

```python
print('new_query_dept')
print(new_query_dept.values)
print('\nExpected')
print(expected.astype(int))
print('\n----')
print(f'chi^2 = {chi2:.4f}')
print(f'pvalue     = {pvalue:.4f}')
```

```python
# null_hypothesis: There is NO relationship between gender and with either Sales or Marketing
# alternate_hypothesis: There IS a relationship gender and with either Sales or Marketing
```

```python
# Conclusion:
# if p < alpha:
#         print('reject the null hypothesis')
# else:
#         print('fail to reject the null hypothesis')
# Answer: fail to reject the null hypothesis
# Answer: there is no relationship between gender and with either Sales or Marketing
```

```python
salary_to_date = pd.crosstab(salary_with_company.from_date, salary_with_company.salary)
salary_to_date
```
